```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime
import time
import json
import re
```

```python
year=range(2013,2021)
month=["01","02","03","04","05","06","07","08","09","10","11","12"]
```

```python
df_all=pd.DataFrame()

for y in year:
    for m in month:
        url="http://www.amac.org.cn//portal/front/statics/fundIndData/findAssetStatics"
        para={
            "year": str(y),
            "month": m
        }

        header={
            "Referer": "http://www.amac.org.cn/researchstatistics/datastatistics/mutualfundindustrydata/",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.81 Safari/537.36 SE 2.X MetaSr 1.0",
            "X-Requested-With": "XMLHttpRequest"
        }
        
        res=requests.get(url,para,headers=header)
        reg1=r'"assetStaticsVOS":(.*?),"curDate"'
        result1=re.findall(reg1,res.text)
        result1=result1[0]
        js_data=json.loads(result1)
        df_data=pd.DataFrame(js_data)
        df_all=df_all.append(df_data)
                
```
